```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "Cleaned_Chronic_Disease_Data_4.csv"
csv_df = pd.read_csv(csv_path)
low_memory=False
csv_df.head()
```

```python
outcsv_df = csv_df[[(x == '%') for x in csv_df['DataValueUnit']]]
```

```python
outcsv_df.head()
```

```python
#output_csv = outcsv_df.to_csv("Cleaned_Chronic_Disease_Data4.csv", index=False, header=True)
outcsv_df2 = outcsv_df[[(x != 2009) for x in outcsv_df['YearStart']]]
outcsv_df2['YearStart'].unique()
```

```python
outcsv_df3 = outcsv_df2.groupby("YearStart")
```

```python
outcsv_2010 = outcsv_df3.get_group(2010)
outcsv_2010.groupby("LocationAbbr")["DataValue"].sum()
```

```python
outcsv_2010_df = pd.DataFrame( outcsv_2010 )
outcsv_2010_df.reset_index()
```

```python
outcsv_2010_df_disability = outcsv_2010_df[[(x == 'Disability') for x in outcsv_2010_df['Topic']]]
#outcsv_df[[(x != 2009) for x in outcsv_df['YearStart']]]
outcsv_2010_df_disability.head()
```

```python
outcsv_2010_df_disability2 = outcsv_2010_df_disability[[(x == 'Overall') for x in outcsv_2010_df_disability['StratificationCategory1']]]
```

```python
x_axis = np.arange(0, (len(outcsv_2010_df_disability2["LocationDesc"])))
plt.title("Density of Disability in States (2010)")
plt.xlabel("State Name")
plt.ylabel("Disability Percentage")
plt.xlim(-0.75, len(outcsv_2010_df_disability2["LocationDesc"]) -.25)
plt.ylim(0, max(outcsv_2010_df_disability2["DataValue"]) + 5)
plt.bar(x_axis, outcsv_2010_df_disability2["DataValue"], facecolor="red", alpha=0.75, align="center")
plt.xticks(x_axis, outcsv_2010_df_disability2["LocationDesc"], rotation = 90)
#plt.figure(figsize=(50,10))
#plt.rcParams['xtick.major.pad']='20'
#plt.set_size_inches(18.5, 10.5)
plt.tight_layout()
plt.savefig('test.png', dpi=400)
plt.show()
```
