```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
GDP_df = pd.read_csv("Resources/Cleaned_GDP_Data.csv")
```

```python
GDP_df.head()
```

```python
group=GDP_df.groupby('State')
Cali=group.get_group('California')
```

```python
Cali2 =Cali.drop(columns=['2017'])
Cali2


Cali2_transposed = Cali2.T 
Cali2_transposed = Cali2_transposed.reset_index() 

new_header = Cali2_transposed.iloc[0]
Cali2_transposed = Cali2_transposed[1:]
Cali2_transposed.columns = new_header
Cali2_transposed
```

```python
group=GDP_df.groupby('State')
Texas=group.get_group('Texas')

Texas2 =Texas.drop(columns=['2017'])
#Texas2

Texas2_transposed = Texas2.T 
Texas2_transposed = Texas2_transposed.reset_index() 
```

```python
new_header = Texas2_transposed.iloc[0]
Texas2_transposed = Texas2_transposed[1:]
Texas2_transposed.columns = new_header
Texas2_transposed
```

```python
group=GDP_df.groupby('State')
Flo=group.get_group('Florida')

Flo2 =Flo.drop(columns=['2017'])

Flo2_transposed = Flo2.T 
Flo2_transposed = Flo2_transposed.reset_index()

new_header = Flo2_transposed.iloc[0]
Flo2_transposed = Flo2_transposed[1:]
Flo2_transposed.columns = new_header
Flo2_transposed

```

```python
group=GDP_df.groupby('State')
NY=group.get_group('New York')

NY2 =NY.drop(columns=['2017'])

NY2_transposed = NY2.T 
NY2_transposed = NY2_transposed.reset_index()

new_header = NY2_transposed.iloc[0]
NY2_transposed = NY2_transposed[1:]
NY2_transposed.columns = new_header
NY2_transposed
```

```python
ax=plt.gca()

Cali2_transposed.plot(kind='line',x='State',y='California',color='red',ax=ax)
Texas2_transposed.plot(kind='line',x='State',y='Texas',color='green',ax=ax)
Flo2_transposed.plot(kind='line',x='State',y='Florida',color='blue',ax=ax)
NY2_transposed.plot(kind='line',x='State',y='New York',color='purple',ax=ax)

plt.title("GDP in Most Populated States")
plt.xlabel("Year")
plt.ylabel("GDP Index")
plt.grid()


plt.show()
plt.savefig("GDP in Most Populated States")
```
